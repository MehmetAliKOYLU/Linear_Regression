```python
import numpy as np
x  = np.array ([[6000],[8200],[9000],[14200],[16200]])
x.reshape(-1,1)
```

```python
y = np.array ([[86000],[82000],[78000],[75000],[70000]])
y.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)

alfa = model.intercept_
beta = model.coef_

print(alfa)
print(beta)
```

```python
test_araba = np.array([[12000]])
tahmini_fiyat = model.predict(test_araba)[0][0]
print(f" araba tahmini fiyati  = {tahmini_fiyat}")
```

```python
y_predictions = model.predict(x)
print(y_predictions)
for i , j in zip(y,y_predictions):
    print("Gercek deger : %.2f til iken tahmin edilen deger = %.2f tl" %(i[0],j[0]))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
MAE = mean_absolute_error(y,y_predictions)
print(MAE)
MSE = mean_squared_error(y,y_predictions)
print(MSE)
R2 = r2_score(y,y_predictions)
print(R2)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.title("Otomobil fiyat-km iliskisi")
plt.xlabel("KM")
plt.ylabel("FIYAT")
plt.plot(x,y,color ='green',label= ('fiyat-km'))
plt.plot(x,y_predictions,color= 'red',label = ('Tahmin Grafigi'))
plt.axis([5000,18000,70000,90000])
plt.grid(True)
```
